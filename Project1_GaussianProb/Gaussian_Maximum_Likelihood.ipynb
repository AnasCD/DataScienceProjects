
MLE of a Gaussian p_model(x|w)




**Explanation of what it does? **
It uses what is known as MLE or Maximum Liklihood Estimation. With the use of MLE as Parameter Estimate, we can find the liklihood for mu or also known as the population/gaussian mean originating from the function 
LL=−N/2log(2πσ^2)−1/2σ^2∑(x_n−μ)2 and finding the partial derivative to be μ = 1/N∑x_n. MLE for variance originates from 
∂LL/∂σ2 = ∂/∂σ2(−N/2log(2πσ^2)−1/2σ^2∑(x_n−μ)2) to produce σ^2 = 1/N∑(x_n−μ)2. Now what does this all mean? - With the use of the data given, it uses these functions in the back end, to find both the parameter variance and mean and produces the gaussian distribution! 

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt1
from scipy.stats import norm

x_axis = np.arange(-20, 30, 0.001)

data = [4, 5, 7, 8, 8, 9, 10, 5, 2, 3, 5, 4, 8, 9]

#Mu as given on Jrmweyer github is 1/N(sum(x_n) 1:N)
#Variance as given on Jrmweyer github is 1/N(sum((x_n-mu)^2) 1:N)
#N is the number of elements in the data

u = 0
for i in data:
  u +=i
u = u/(len(data))

#var = 0
#for r in data:
  #var += (i - u)**2

#var = var/(len(data))

var = np.var(data) #Instead of the above code, we can use numpy structure to get the variance



plt.plot(x_axis, norm.pdf(x_axis,u,var),"g-")
plt.xlabel("X")
plt.ylabel("Gaussian Maximum Likelihood")
plt.grid()

plt.show()

print("Variance: " ,var)
print("μ: " , u)


```

MLE of a conditional Gaussian p_model(y|x,w)

This is a gradient descent algorithm use for a linear regression model of best fit! The functionality of gradient descent lets us find the right "M" and "B" to precisely, with low mean squared error to calculate the best fit linear regression model for the x values and their y outputs! With the use of partial derivatives, this allows us to iterate through learning rates and produces "m_current" and "b_current" corresponding to y = mx + b function. The multiple lines on the plot are steps to getting a closer linear regression model with its learning rate and derivative. 

```python
import numpy as np
import matplotlib.pyplot as plt

#Lets define a function that makes use of gradient Descent Algorithm
#Cost Function(Mean squared Error) -> (1/n)*sum(y-y(hat))^2
#Partial Derivative with respect to m -> -(2/n)*x*sum(y-y(hat))
#Partial Derivative with respect to b -> -(2/n)*sum(y-y(hat))
#Steps m = m - leaningRate*d/dm and b = b - learningRate*d/db

def gradient_descent_Algo(x,y):
  m_current = 0
  b_current = 0
  iterations = 10000 #The more iterations, the closer you get to a best fit line!(Change this if needed)
  learningRate = 0.00001 #This changes depending on the best fit line
  for i in range(iterations):
    y_hat = m_current*x+b_current #This is the form y = mx + b concidering that best fit line is a linear model
    f_Difference = y - y_hat
    cost = (1/len(x))*sum([value**2 for value in (f_Difference)])#This is the cost function presented above (Line 5)

    

    m_derivative = -(2/len(x))*sum(x*(f_Difference))#Derivative with respect to m
    b_derivative = -(2/len(x))*sum(f_Difference)#Derivative with respect to b
    
    m_current = m_current - learningRate*m_derivative
    b_current = b_current - learningRate*b_derivative

    #Uncomment the below print function to see m, b, cost(Error), and iteration Number
    #print("m {}, b {},cost {} iteration {}".format(m_current,b_current,cost,i))
    plt.plot(x,y_hat)

  print("Current M~ {}, Current B~ {}".format(m_current, b_current))

  


x_values = [8, 16, 22, 33, 50, 51]
y_values = [5, 20, 14, 32, 42, 58]
x = np.array(x_values)
y = np.array(y_values)
plt.plot(x_values,y_values, 's',marker="o",markersize = 14,color = "red")
gradient_descent_Algo(x,y)

```
